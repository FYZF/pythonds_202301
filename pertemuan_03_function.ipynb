```python
import pyttsx3
```

```python
engine = pyttsx3.init('nsss')
```

```python
voices = engine.getProperty('voices')
```

```python
engine.setProperty('voices', voices[1].id)
```

```python
def bicara(text):
    engine.say(text)
    engine.runAndWait()
```

```python
bicara("Haloo Apa kabar")
```
